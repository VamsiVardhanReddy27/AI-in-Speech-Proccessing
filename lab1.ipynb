# QUESTION 01:

```python
#  A1. Load the recorded speech file into your python workspace. 
#Once loaded, plot the graph for the speech signal. 
#You may use the below code from librosa as a reference
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
audiofile = r"C:\Users\vamsi\Desktop\AI in speech enhancement\lab\lab1\Recording.wav"

# Load the speech file
y, sr = librosa.load(audiofile)
print(y)
print(sr) # sampling rate of y
plt.figure(figsize=(16, 6))
librosa.display.waveshow(y)
plt.title('Speech Signal')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.show()
```

# Question: 02

```python
# A2. Observe the length & magnitude range of the signal from the plot.
# Observing the plot, relate to the spoken words and silence between the words.
```

Length of the Signal (Time Axis):

The length of the signal is the overall duration of the signal is from the beginning to the end of the plot.

Magnitude Range (Amplitude Axis):

The magnitude range is the difference between the highest and lowest points on the y-axis.

Relation to Spoken Words and Silence:

Spoken words are typically characterized by higher amplitudes, resulting in peaks in the plot.
Silence or pauses between words are represented by regions of lower amplitude.

```python
duration = len(y) / sr  # Duration in seconds
print(f"Duration of the audio: {duration:.2f} seconds")
print(f"Magnitude range: {y.min()} to {y.max()}")
```

# Question: 03

```python
# A3. Take a small segment of the signal and play it. 
```

```python
import IPython.display as ipd

# Load the speech file
y, sr = librosa.load(audiofile)

start_time1 = 1
end_time1 = 2  
segment1 = y[int(start_time1 * sr):int(end_time1 * sr)]

# Play the segment
ipd.Audio(segment1, rate=sr)
```
